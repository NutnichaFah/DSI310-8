### read stored data from DataPlatform

```python
from dsmlibrary.datanode import DataNode
from tqdm.auto import tqdm

import json
import pandas as pd
```

```python
dir_raw_id = 62
dir_process_id = 63
```

```python
datanode = DataNode()
```

```python
meta, fp = datanode.get_file(datanode.get_file_id(directory_id=dir_raw_id, name='jewelry_events_top100.json'))
meta
```

```python
data = json.load(fp)
data
```

```python
df = pd.json_normalize(data)
```

### write DataNode to DataPlatform

```python
datanode.write( df=df, 
                directory=dir_process_id, 
                name="Jewelry_events_top100", 
                profiling=True, 
                lineage=[datanode.get_file_id(directory_id=dir_raw_id, name='jewelry_events_top100.json')],
              )
```
