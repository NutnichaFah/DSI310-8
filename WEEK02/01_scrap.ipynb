# Import Library

```python
import scrapy
import pandas as pd
import json
import csv

from datetime import datetime
from scrapy.crawler import CrawlerProcess
from scrapy.crawler import CrawlerRunner
```

```python
class wk2Spider(scrapy.Spider):
    name = "jew_news"
    # allowed_domains = ['economictimes.indiatimes.com']
    start_urls = ['https://economictimes.indiatimes.com/topic/jewellery-industry/',]
    
    custom_settings = {
        'FEEDS': { 'jew_news.csv': { 'format': 'csv',}}
        }
    
    def parse(self, response):
        for news in response.css('div.clr.flt.topicstry.story_list'):
            yield{
            'Topic': news.css('div.contentD h2::text').get().replace("\u2019", " "),
            'Describe' : news.css('div.syn::text').get().replace("\u2019", " ").replace("\"", " ").replace("\u201c", " ").replace("\u201d", " "),
            'Datetime': news.css('time::text').get(),
            'source_url' : news.css('div.contentD a::attr(href)').get() 
        }
```

```python
process = CrawlerProcess()
process.crawl(wk2Spider)
process.start()
```

```python
df=pd.read_csv('jew_news.csv')
df
```
